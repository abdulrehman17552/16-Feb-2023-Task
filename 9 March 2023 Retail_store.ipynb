```python
df['Total Price'] = df['Quantity']* df['Price']
df.tail()
```

```python
df['Country'].unique()
```

```python
m= df.groupby('Country')['Total Price'].sum().sort_values(ascending = False)[:5].plot(kind='bar')
m
```

#### According to the Above graph United Kingdom is the country where people are spending more. while Netherlands, Ireland, Germeny and France are potential places which can be targeted.

```python
df.groupby('Country')['Quantity'].count().sort_values(ascending = False)[:5].plot(kind='bar')
```

```python

df['Month'] =  df['InvoiceDate'].dt.month
df.head()
```

```python
df.groupby('Month')['Total Price'].sum().sort_values(ascending = False).plot(kind='bar', color= 'red')
```

**In two years of records we found that the last quarter has more revenues among others while november is at the top**
